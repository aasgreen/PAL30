# Data Analysis for PAL30 Voltage Switching

Temperature: 75 degC

Notes: We see weird undulations at layer boundaries, but no obvious switching effects.

```python
%matplotlib inline #Inline plotting with sensible defaults
pylab.rcParams['figure.figsize'] = 9, 6
pylab.rcParams['font.size'] = 18
```

### Import Data

```python
D1 = np.loadtxt("../AppliedVoltage/meangreyappliedvoltage.txt",unpack=True)
rotateD2 = np.loadtxt("../AppliedVoltage/RotatedStage/Values.txt",unpack=True)
control = np.loadtxt("./controlfromT98Cooling.txt",unpack=True)
```

### Define Global Variables

```python
#tstep = 1/fps
tstep = 1./9.36  #in seconds/frame
```

### Initial Plots of Data Normalized so Mean Grey Value = 1

```python
plt.plot(D1[0]*tstep,D1[1]/np.mean(D1[1]))
plt.plot(rotateD2[0]*tstep,rotateD2[1]/np.mean(rotateD2[1]))
plt.xlabel("Time (S)")
plt.ylabel("Mean Grey Value")
```

### Fourier Analysis

We want to extract the frequency of the oscillations in the data. Unfortunately, I didn't think to take a control (no voltage applied), so we can't be sure if the oscillations are due to the light source flucuatings. However, I will use some additional data where video was taken to try to rule out light source oscillations.

```python
fdat1 = np.fft.fft(D1[1])
freq1 = np.fft.fftfreq(D1[0].shape[-1], d=tstep)
s = fdat1.shape[-1]
fdat1 = fdat1/s #normalize

fdat2 = np.fft.fft(rotateD2[1])
freq2 = np.fft.fftfreq(rotateD2[0].shape[-1], d=tstep)
s2 = fdat2.shape[-1]
fdat2 = fdat2/s2 #normalize

fdatcon = np.fft.fft(control[1])
freqcon = np.fft.fftfreq(control[0].shape[-1], d=tstep)
scon = fdatcon.shape[-1]
fdatcon = fdatcon/scon #normalize
```

```python
plt.plot(freq1[:s/2],np.abs(fdat1[:s/2]),label='Initial')
plt.ylim([0,10])

plt.plot(freq2[:s2/2],np.abs(fdat2[:s2/2]),label='Stage Rot. 30 deg')


plt.plot(freqcon[:scon/2],np.abs(fdatcon[:scon/2]), label='Control (No V)')

plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude Unnormalized")
plt.legend(loc='best')

plt.tight_layout()
plt.savefig("initialresults.png")
```

As we see the same peak at around 1.8 Hz, we can't conclude that there is any effect going on. The next thing to try will be to look varying the frequency to get out of the noisy 2 Hz range, and see if this effect goes as the temperature is changed.
