### Import Data

### Define Global Variables

```python
#tstep = 1/fps
tstep = 1./7  #in seconds/frame
```

### Initial Plots of Data Normalized so Mean Grey Value = 1

```python
plt.plot(D1[0]*tstep,D1[1]/np.mean(D1[1]))
plt.xlabel("Time (S)")
plt.ylabel("Mean Grey Value")
```

### Fourier Analysis

We want to extract the frequency of the oscillations in the data. Unfortunately, I didn't think to take a control (no voltage applied), so we can't be sure if the oscillations are due to the light source flucuatings. However, I will use some additional data where video was taken to try to rule out light source oscillations.

```python
fdat1 = np.fft.fft(D1[1])
freq1 = np.fft.fftfreq(D1[0].shape[-1], d=tstep)
s = fdat1.shape[-1]
fdat1 = fdat1/s #normalize
```

```python
plt.plot(freq1[:s/2],np.abs(fdat1[:s/2]),label='Initial')
plt.ylim([0,10])

plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude Unnormalized")
plt.legend(loc='best')

plt.ylim([0,2])

plt.tight_layout()
plt.savefig("initialresults.png")
```
