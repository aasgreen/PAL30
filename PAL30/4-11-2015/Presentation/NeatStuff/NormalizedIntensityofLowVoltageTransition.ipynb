```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import glob
from scipy.optimize import curve_fit
from matplotlib import rcParams
rcParams['axes.labelsize'] = 32
rcParams['xtick.labelsize'] = 30
rcParams['ytick.labelsize'] = 30
rcParams['legend.fontsize'] = 30
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = True
rcParams['figure.figsize'] = (9.0, 4.5)
```

```python
data = np.loadtxt('zxis-regionoffilm-heatedwith5Von-81-102-wholeimage.dat',unpack=True)
```

```python
#Now, fit data
def fit(x,a,b,c,d,e,f,g,h):
    return a+(b*x+c*x**3+d*x**4+e*x**5+f*x**6+g*x**7+h*x**8)
```

```python
popt,pcov = curve_fit(fit,data[0],data[1])
#plt.plot(data[0],data[1],label='Intensity')
plt.plot(data[0],fit(data[0],*popt)-data[1])
```

```python
plt.plot(data[0],fit(data[0],*popt)-data[1],label='Normalized Intensity')
plt.xlabel(r'Frame')
plt.ylabel(r'Intensity $(Arb. Units)$')
plt.legend(loc='best')
plt.tight_layout(pad=.4)

plt.savefig("intensityoft81-t102-5Vpp.png")
```
