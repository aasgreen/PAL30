```python
#show scott a simple transfer function implementation:
fnoise_approx = fred.data[1]**2./(fred.aveV/2*np.sqrt(2)*np.pi*fred.delay)**2
```

```python
plt.subplot(211)
plt.loglog(fred.fnoise[0],fred.fnoise[1],label="Exact Transfer")
plt.loglog(fred.fnoise[0],fnoise_approx,label="Taylor Expand")
plt.legend(loc='best')
plt.xlabel("Hz")
plt.ylabel(r"$\frac{\textrm{Hz}^2}{\textrm{Hz}}$")
plt.subplot(212)
plt.loglog(fred.fnoise[0],er,label=r"$\frac{\textrm{Exact}-\textrm{Taylor}}{\textrm{Exact}}$")
plt.xlabel("Hz")
plt.ylabel("Relative Error")
plt.tight_layout(pad=.4)
plt.legend(loc='best')
plt.savefig("approxtransfer.pdf")
plt.show()
```
